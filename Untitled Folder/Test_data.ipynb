```python
import pandas as pd
import numpy as np
df = pd.read_csv('CBC_data.csv')
df
df.drop_duplicates(inplace=True)
df.drop([365], inplace=True)
df.drop([0], inplace=True)
newdf=df.astype('float64')
# newdf dataframe allows to change the data type of all columns 
newdf.info()
newdf.describe()
#getting rid of the white spaces in the columns names 
newdf.columns = newdf.columns.str.replace(' ', '')
newdf.columns
newdf.describe()
pd.set_option('display.max_column', None)
pd.set_option('display.max_rows', None)
#newdf['Age'].value_counts(ascending=True)
# this shows the number of kids under 15 ------> 14 kids
len(newdf.Age[newdf.Age<15])
# The columns with kids younger than 15 would need to be dropped as Anemia's parameters normal range differs 
newdf.drop(newdf[newdf['Age'] < 15].index,inplace=True)
newdf.reset_index(inplace=True)
newdf.drop(newdf[newdf['Age'] < 15].index,inplace=True)
newdf.drop(['S.No.', 'index'], axis=1, inplace=True)
conditions = [
    (newdf['Sex'] == 0.0) & (newdf['HGB'] < 13) & (newdf['MCV'] < 80),
    (newdf['Sex'] == 0.0) & (newdf['HGB'] < 13) & (newdf['MCV'] > 100),
    (newdf['Sex'] == 0.0) & ((newdf['HGB'] < 13) & ((newdf['MCV'] >= 80) & (newdf['MCV'] <= 100))),
    (newdf['Sex'] == 1.0) & (newdf['HGB'] < 12) & (newdf['MCV'] < 80),
    (newdf['Sex'] == 1.0) & (newdf['HGB'] < 12) & (newdf['MCV'] > 100),
    (newdf['Sex'] == 1.0) & ((newdf['HGB'] < 12) & ((newdf['MCV'] >= 80) & (newdf['MCV'] <= 100)))
]
values = [1, 3, 2, 1, 3, 2]

# Create the result column based on the conditions
newdf['result'] = np.select(conditions, values, default=0)

newdf
```

```python
from imblearn.over_sampling import SMOTE
# copying the newdf dataframe to then balance it 
df_to_balance=newdf.copy()
X = df_to_balance.drop('result', axis=1)
y = df_to_balance['result']

# Apply SMOTE to balance the dataset
smote = SMOTE()
X_balanced, y_balanced = smote.fit_resample(X, y)
X_balanced_df=pd.DataFrame(X_balanced)
y_balanced_df=pd.DataFrame(y_balanced)
#X_balanced_df
df_after_balance = pd.concat([X_balanced_df, y_balanced_df], axis=1)
df_after_balance=df_after_balance.round(2)
df_after_balance['Age'] = df_after_balance['Age'].astype(int)
df_after_balance['Sex'] = df_after_balance['Sex'].astype(int)
X_balanced = df_after_balance.drop('result', axis=1)
y_balanced = df_after_balance['result']
X_balanced
```

```python

import matplotlib.pyplot as plt
from sklearn import svm
from itertools import combinations

# Extract features (X) and target variable (y) from the dataframe
X = X_balanced.values
y = y_balanced.values
print(X.shape)
print(y.shape)

# Get all combinations of feature indices
feature_indices = range(X.shape[1])
feature_combinations = list(combinations(feature_indices, 2))

# Create subplots for each feature combination
fig, axs = plt.subplots(len(feature_combinations), figsize=(10, 10))

# Fit SVM model and plot decision boundary for each feature combination
for i, (feat1, feat2) in enumerate(feature_combinations):
    # Fit SVM model
    clf = svm.SVC(kernel='linear')
    clf.fit(X[:, [feat1, feat2]], y)

    # Get support vectors and plot data points
    support_vectors = clf.support_vectors_
    axs[i].scatter(X[:, feat1], X[:, feat2], c=y)
    axs[i].scatter(support_vectors[:, 0], support_vectors[:, 1], color='red', marker='x')

    # Plot the decision boundary
    ax = axs[i]
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # Create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T

    Z = clf.predict(xy)
    Z = Z.reshape(XX.shape)
    # Plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
    ax.set_xlabel(f"Feature {feat1+1}")
    ax.set_ylabel(f"Feature {feat2+1}")
    ax.set_title(f"Linear SVM Decision Boundary for Features {feat1+1} and {feat2+1}")

plt.tight_layout()
plt.show()
```
