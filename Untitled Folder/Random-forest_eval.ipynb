```python
import pandas as pd
import numpy as np
df = pd.read_csv('CBC_data.csv')
df
```

```python
df.drop_duplicates(inplace=True)
df.drop([365], inplace=True)
df.drop([0], inplace=True)
newdf=df.astype('float64')
# newdf dataframe allows to change the data type of all columns 
newdf.info()
newdf.describe()
#getting rid of the white spaces in the columns names 
newdf.columns = newdf.columns.str.replace(' ', '')
newdf.columns
newdf.describe()
pd.set_option('display.max_column', None)
pd.set_option('display.max_rows', None)
#newdf['Age'].value_counts(ascending=True)
# this shows the number of kids under 15 ------> 14 kids
len(newdf.Age[newdf.Age<15])
# The columns with kids younger than 15 would need to be dropped as Anemia's parameters normal range differs 
newdf.drop(newdf[newdf['Age'] < 15].index,inplace=True)
newdf.reset_index(inplace=True)
newdf.drop(newdf[newdf['Age'] < 15].index,inplace=True)
newdf
```

```python
newdf.drop(['S.No.', 'index'], axis=1, inplace=True)
newdf
```

```python
conditions = [
    (newdf['Sex'] == 0.0) & (newdf['HGB'] < 13) & (newdf['MCV'] < 80),
    (newdf['Sex'] == 0.0) & (newdf['HGB'] < 13) & (newdf['MCV'] > 100),
    (newdf['Sex'] == 0.0) & ((newdf['HGB'] < 13) & ((newdf['MCV'] >= 80) & (newdf['MCV'] <= 100))),
    (newdf['Sex'] == 1.0) & (newdf['HGB'] < 12) & (newdf['MCV'] < 80),
    (newdf['Sex'] == 1.0) & (newdf['HGB'] < 12) & (newdf['MCV'] > 100),
    (newdf['Sex'] == 1.0) & ((newdf['HGB'] < 12) & ((newdf['MCV'] >= 80) & (newdf['MCV'] <= 100)))
]
values = [1, 3, 2, 1, 3, 2]

# Create the result column based on the conditions
newdf['result'] = np.select(conditions, values, default=0)

newdf
```

```python
from imblearn.over_sampling import SMOTE
# copying the newdf dataframe to then balance it 
df_to_balance=newdf.copy()
X = df_to_balance.drop('result', axis=1)
y = df_to_balance['result']

# Apply SMOTE to balance the dataset
smote = SMOTE()
X_balanced, y_balanced = smote.fit_resample(X, y)
X_balanced_df=pd.DataFrame(X_balanced)
y_balanced_df=pd.DataFrame(y_balanced)
#X_balanced_df
df_after_balance = pd.concat([X_balanced_df, y_balanced_df], axis=1)
df_after_balance=df_after_balance.round(2)
df_after_balance['Age'] = df_after_balance['Age'].astype(int)
df_after_balance['Sex'] = df_after_balance['Sex'].astype(int)
X_balanced = df_after_balance.drop('result', axis=1)
y_balanced = df_after_balance['result']
X_balanced

```

```python
values = X_balanced['Sex'].value_counts()
values
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score, train_test_split
from sklearn.feature_selection import SelectFromModel

X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.2, random_state=42)



# Define the parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [10,20,40,60,80,100,200,250,300],
    'max_depth': [None, 5, 10],
    'max_features': [None, 'sqrt', 'log2'],
    'criterion': ['gini', 'entropy']
}

# Create the Random Forest classifier
rf = RandomForestClassifier()


# Perform hyperparameter tuning using GridSearchCV with 10-fold cross-validation
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=10)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Create a new Random Forest classifier with the best hyperparameters
best_rf = RandomForestClassifier(**best_params)

# Perform feature selection
selector = SelectFromModel(best_rf, max_features=5)
selector.fit(X_train, y_train)

# Get the selected feature indices
selected_feature_indices = selector.get_support(indices=True)

# Get the selected feature names
selected_feature_names = [ index for index in selected_feature_indices]

# Print the selected feature names
print("Selected Features:")
for feature_name in selected_feature_names:
    print(feature_name)

# Fit the classifier on the training data
XTRAIN=pd.DataFrame(X_train)

XTEST=pd.DataFrame(X_test)
#print(XTEST.iloc[:,selected_feature_names])
best_rf.fit(XTRAIN.iloc[:,selected_feature_names], y_train)

# Calculate the training data accuracy for each fold using cross-validation
train_accuracy_scores = cross_val_score(best_rf, XTRAIN.iloc[:,selected_feature_names], y_train, cv=10)

# Calculate the testing data accuracy
test_accuracy = best_rf.score(XTEST.iloc[:,selected_feature_names], y_test)

# Print the results
print("Best Hyperparameters:", best_params)
print("Training Data Accuracy for each fold:")
for fold, accuracy in enumerate(train_accuracy_scores):
    print(f"Fold {fold+1}: {accuracy}")
print("Testing Data Accuracy:", test_accuracy)
```

```python
from sklearn.metrics import confusion_matrix, recall_score, f1_score, precision_score, classification_report
y_pred = best_rf.predict(XTEST.iloc[:, selected_feature_names])

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Calculate the recall, F1 score, and precision
f1 = f1_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')

# Print the results
print("Best Hyperparameters:", best_params)
print("Training Data Accuracy for each fold:")
for fold, accuracy in enumerate(train_accuracy_scores):
    print(f"Fold {fold+1}: {accuracy}")
print("Testing Data Accuracy:", test_accuracy)
print("Confusion Matrix:")
print(cm)
print("Recall:", recall)
print("F1 Score:", f1)
print("Precision:", precision)
classification_rep = classification_report(y_test, y_pred)
print("Classification Report:")
print(classification_rep)
```
