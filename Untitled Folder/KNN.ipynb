```python
import pandas as pd
import numpy as np
df = pd.read_csv('CBC_data.csv')
df
df.drop_duplicates(inplace=True)
df.drop([365], inplace=True)
df.drop([0], inplace=True)
newdf=df.astype('float64')
# newdf dataframe allows to change the data type of all columns 
newdf.info()
newdf.describe()
#getting rid of the white spaces in the columns names 
newdf.columns = newdf.columns.str.replace(' ', '')
newdf.columns
newdf.describe()
pd.set_option('display.max_column', None)
pd.set_option('display.max_rows', None)
#newdf['Age'].value_counts(ascending=True)
# this shows the number of kids under 15 ------> 14 kids
len(newdf.Age[newdf.Age<15])
# The columns with kids younger than 15 would need to be dropped as Anemia's parameters normal range differs 
newdf.drop(newdf[newdf['Age'] < 15].index,inplace=True)
newdf.reset_index(inplace=True)
newdf.drop(newdf[newdf['Age'] < 15].index,inplace=True)
newdf.drop(['S.No.', 'index'], axis=1, inplace=True)
conditions = [
    (newdf['Sex'] == 0.0) & (newdf['HGB'] < 13) & (newdf['MCV'] < 80),
    (newdf['Sex'] == 0.0) & (newdf['HGB'] < 13) & (newdf['MCV'] > 100),
    (newdf['Sex'] == 0.0) & ((newdf['HGB'] < 13) & ((newdf['MCV'] >= 80) & (newdf['MCV'] <= 100))),
    (newdf['Sex'] == 1.0) & (newdf['HGB'] < 12) & (newdf['MCV'] < 80),
    (newdf['Sex'] == 1.0) & (newdf['HGB'] < 12) & (newdf['MCV'] > 100),
    (newdf['Sex'] == 1.0) & ((newdf['HGB'] < 12) & ((newdf['MCV'] >= 80) & (newdf['MCV'] <= 100)))
]
values = [1, 3, 2, 1, 3, 2]

# Create the result column based on the conditions
newdf['result'] = np.select(conditions, values, default=0)

newdf
```

```python
from imblearn.over_sampling import SMOTE
# copying the newdf dataframe to then balance it 
df_to_balance=newdf.copy()
X = df_to_balance.drop('result', axis=1)
y = df_to_balance['result']

# Apply SMOTE to balance the dataset
smote = SMOTE()
X_balanced, y_balanced = smote.fit_resample(X, y)
X_balanced_df=pd.DataFrame(X_balanced)
y_balanced_df=pd.DataFrame(y_balanced)
#X_balanced_df
df_after_balance = pd.concat([X_balanced_df, y_balanced_df], axis=1)
df_after_balance=df_after_balance.round(2)
df_after_balance['Age'] = df_after_balance['Age'].astype(int)
df_after_balance['Sex'] = df_after_balance['Sex'].astype(int)
X_balanced = df_after_balance.drop('result', axis=1)
y_balanced = df_after_balance['result']
X_balanced
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score, ShuffleSplit
from sklearn.feature_selection import RFE
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_balanced_scaled = scaler.fit_transform(X_balanced)

X_train, X_test, y_train, y_test = train_test_split(X_balanced_scaled, y_balanced, test_size=0.2, random_state=42)

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto'],
    'p': [1, 2]
}

# Create the KNN classifier
knn = KNeighborsClassifier()

# Perform hyperparameter tuning using GridSearchCV with 10-fold cross-validation
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=10)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Create a new KNN classifier with the best hyperparameters
best_knn = KNeighborsClassifier(**best_params)

# Fit the model on the training data
best_knn.fit(X_train, y_train)

shuffle_split = ShuffleSplit(n_splits=10, test_size=0.2, random_state=42)

# Calculate the training data accuracy for each fold using cross-validation
train_accuracy_scores = cross_val_score(best_knn, X_train, y_train, cv=shuffle_split)

# Calculate the testing data accuracy
test_accuracy = best_knn.score(X_test, y_test)

# Print the results
print("Best Hyperparameters:", best_params)
print("Training Data Accuracy for each fold:")
for fold, accuracy in enumerate(train_accuracy_scores):
    print(f"Fold {fold+1}: {accuracy}")
print("Testing Data Accuracy:", test_accuracy)
```

### using ANOVA with the select best

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score, ShuffleSplit
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_balanced_scaled = scaler.fit_transform(X_balanced)

X_train, X_test, y_train, y_test = train_test_split(X_balanced_scaled, y_balanced, test_size=0.2, random_state=42)

# Perform feature selection using ANOVA
feature_selector = SelectKBest(score_func=f_classif, k=6)
X_train_selected = feature_selector.fit_transform(X_train, y_train)
X_test_selected = feature_selector.transform(X_test)

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'n_neighbors':  [3,4, 5,6, 7,8, 9,10,12,15],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}

# Create the KNN classifier
knn = KNeighborsClassifier()

# Perform hyperparameter tuning using GridSearchCV with 10-fold cross-validation
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=10)
grid_search.fit(X_train_selected, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Create a new KNN classifier with the best hyperparameters
best_knn = KNeighborsClassifier(**best_params)

best_knn.fit(X_train_selected, y_train)
shuffle_split = ShuffleSplit(n_splits=10, test_size=0.2, random_state=42)
# Calculate the training data accuracy for each fold using cross-validation
train_accuracy_scores = cross_val_score(best_knn, X_train_selected, y_train, cv=shuffle_split)

# Calculate the testing data accuracy
test_accuracy = best_knn.score(X_test_selected, y_test)

# Print the results
print("Best Hyperparameters:", best_params)
print("Training Data Accuracy for each fold:")
for fold, accuracy in enumerate(train_accuracy_scores):
    print(f"Fold {fold+1}: {accuracy}")
print("Testing Data Accuracy:", test_accuracy)
```

```python
selected_feature_names = [X_balanced.columns[i] for i in feature_selector.get_support(indices=True)]
print("Selected Features:", selected_feature_names)
```

```python
from sklearn.metrics import confusion_matrix, recall_score, f1_score, precision_score, classification_report
y_pred = best_knn.predict(X_test_selected)

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Calculate the recall, F1 score, and precision
f1 = f1_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')

# Print the results
print("Best Hyperparameters:", best_params)
print("Training Data Accuracy for each fold:")
for fold, accuracy in enumerate(train_accuracy_scores):
    print(f"Fold {fold+1}: {accuracy}")
print("Testing Data Accuracy:", test_accuracy)
print("Confusion Matrix:")
print(cm)
print("Recall:", recall)
print("F1 Score:", f1)
print("Precision:", precision)
classification_rep = classification_report(y_test, y_pred)
print("Classification Report:")
print(classification_rep)
```
