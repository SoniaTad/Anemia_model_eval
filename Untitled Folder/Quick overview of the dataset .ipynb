```python
import pandas as pd
import numpy as np
df = pd.read_csv('CBC_data.csv')
df
```

```python
df.drop_duplicates(inplace=True)
df.drop([365], inplace=True)
df.drop([0], inplace=True)
newdf=df.astype('float64')
# newdf dataframe allows to change the data type of all columns 
newdf.info()
newdf.describe()
#getting rid of the white spaces in the columns names 
newdf.columns = newdf.columns.str.replace(' ', '')
newdf.columns
newdf.describe()
pd.set_option('display.max_column', None)
pd.set_option('display.max_rows', None)
#newdf['Age'].value_counts(ascending=True)
# this shows the number of kids under 15 ------> 14 kids
len(newdf.Age[newdf.Age<15])
# The columns with kids younger than 15 would need to be dropped as Anemia's parameters normal range differs 
newdf.drop(newdf[newdf['Age'] < 15].index,inplace=True)
newdf.reset_index(inplace=True)
newdf.drop(newdf[newdf['Age'] < 15].index,inplace=True)
newdf
```

```python
newdf.drop(['S.No.', 'index'], axis=1, inplace=True)
newdf
```

```python
print(sum(newdf['Sex'] == 1.0))
```

```python
conditions = [
    (newdf['Sex'] == 0.0) & (newdf['HGB'] < 13) & (newdf['MCV'] < 80),
    (newdf['Sex'] == 0.0) & (newdf['HGB'] < 13) & (newdf['MCV'] > 100),
    (newdf['Sex'] == 0.0) & ((newdf['HGB'] < 13) & ((newdf['MCV'] >= 80) & (newdf['MCV'] <= 100))),
    (newdf['Sex'] == 1.0) & (newdf['HGB'] < 12) & (newdf['MCV'] < 80),
    (newdf['Sex'] == 1.0) & (newdf['HGB'] < 12) & (newdf['MCV'] > 100),
    (newdf['Sex'] == 1.0) & ((newdf['HGB'] < 12) & ((newdf['MCV'] >= 80) & (newdf['MCV'] <= 100)))
]
values = [1, 3, 2, 1, 3, 2]

# Create the result column based on the conditions
newdf['result'] = np.select(conditions, values, default=0)

newdf
```

```python
print(sum(newdf['result'] == 0))
print(sum(newdf['result'] == 2))
print(sum(newdf['result'] == 3))
```

```python
result = newdf[(newdf['result'].isin([1, 2, 3])) & (newdf['Sex'] == 0.0)].count()
print(result)
```

```python
from imblearn.over_sampling import SMOTE
# copying the newdf dataframe to then balance it 
df_to_balance=newdf.copy()
X = df_to_balance.drop('result', axis=1)
y = df_to_balance['result']

# Apply SMOTE to balance the dataset
smote = SMOTE()
X_balanced, y_balanced = smote.fit_resample(X, y)
X_balanced_df=pd.DataFrame(X_balanced)
y_balanced_df=pd.DataFrame(y_balanced)
# Combine the balanced dataframes
combined_df = pd.concat([X_balanced_df, y_balanced_df], axis=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the age ranges
age_bins = [16, 26, 41, 66, 100]

# Create the age range labels
age_labels = ['16.0-25.0', '26.0-40.0', '41.0-65.0', '66.0-100.0']

# Categorize the ages into the defined bins
newdf['AgeRange'] = pd.cut(newdf['Age'], bins=age_bins, labels=age_labels, right=False)

# Create separate dataframes based on each result and sex
result_0_sex_1 = newdf[(newdf['result'] == 0) & (newdf['Sex'] == 1.0)]
result_0_sex_0 = newdf[(newdf['result'] == 0) & (newdf['Sex'] == 0.0)]
result_1_sex_1 = newdf[(newdf['result'] == 1) & (newdf['Sex'] == 1.0)]
result_1_sex_0 = newdf[(newdf['result'] == 1) & (newdf['Sex'] == 0.0)]
result_2_sex_1 = newdf[(newdf['result'] == 2) & (newdf['Sex'] == 1.0)]
result_2_sex_0 = newdf[(newdf['result'] == 2) & (newdf['Sex'] == 0.0)]
result_3_sex_1 = newdf[(newdf['result'] == 3) & (newdf['Sex'] == 1.0)]
result_3_sex_0 = newdf[(newdf['result'] == 3) & (newdf['Sex'] == 0.0)]

# Create the bar chart
fig, ax = plt.subplots()
bar_width = 0.15
index = np.arange(len(age_labels))

bar1 = ax.bar(index, [len(result_0_sex_1[result_0_sex_1['AgeRange'] == label]) for label in age_labels], bar_width, label='Result 0, Sex 1.0')
bar2 = ax.bar(index + bar_width, [len(result_0_sex_0[result_0_sex_0['AgeRange'] == label]) for label in age_labels], bar_width, label='Result 0, Sex 0.0')
bar3 = ax.bar(index + 2*bar_width, [len(result_1_sex_1[result_1_sex_1['AgeRange'] == label]) for label in age_labels], bar_width, label='Result 1, Sex 1.0')
bar4 = ax.bar(index + 3*bar_width, [len(result_1_sex_0[result_1_sex_0['AgeRange'] == label]) for label in age_labels], bar_width, label='Result 1, Sex 0.0')
bar5 = ax.bar(index + 4*bar_width, [len(result_2_sex_1[result_2_sex_1['AgeRange'] == label]) for label in age_labels], bar_width, label='Result 2, Sex 1.0')
bar6 = ax.bar(index + 5*bar_width, [len(result_2_sex_0[result_2_sex_0['AgeRange'] == label]) for label in age_labels], bar_width, label='Result 2, Sex 0.0')
bar7 = ax.bar(index + 6*bar_width, [len(result_3_sex_1[result_3_sex_1['AgeRange'] == label]) for label in age_labels], bar_width, label='Result 3, Sex 1.0')
bar8 = ax.bar(index + 7*bar_width, [len(result_3_sex_0[result_3_sex_0['AgeRange'] == label]) for label in age_labels], bar_width, label='Result 3, Sex 0.0')

ax.set_xlabel('Age Range')
ax.set_ylabel('Count')
ax.set_title('Count of Age Ranges by Result and Sex')
ax.set_xticks(index + 3.5*bar_width)
ax.set_xticklabels(age_labels)
ax.legend()

plt.show()
```
